**Data Analytics Methods Using R**

```r
# Create numeric vectors using c()
height <- c(150, 160, 170, 155, 165, 175, 180)
weight <- c(45, 50, 65, 48, 55, 70, 75)
group  <- c("A", "A", "B", "B", "C", "C", "C")   # categorical variable

```

**Descriptive Statistics**

```r

mean(height)
median(height)
sd(height)
var(height)
range(height)

# Correlation between height and weight
cor(height, weight)
```


**Statistical Methods for Evaluation**


```r

# t-test between two groups (A and B)
height_A <- height[group == "A"]
height_B <- height[group == "B"]
t.test(height_A, height_B)
```

**Hypothesis Testing ---- ANOVA**


```r
anova_model <- aov(height ~ group)
summary(anova_model)
```

**Hypothesis Testing ---- Chi-squared test**

```r
# Suppose a die was rolled 60 times, and we observed:
observed <- c(8, 9, 10, 11, 12, 10)     # observed counts
expected <- c(10, 10, 10, 10, 10, 10)   # expected (equal probabilities)

# Perform Chi-Square Goodness of Fit Test
chisq.test(x = observed, p = expected / sum(expected))

# -------------------------------------------
# Example 2: Test of Independence
# -------------------------------------------
# Create a 2x2 contingency table:
# Example: Gender vs Product Preference

gender <- c("Male", "Male", "Female", "Female")
product <- c("A", "B", "A", "B")
frequency <- c(30, 20, 25, 25)

# Create the table
table_data <- matrix(frequency,
                     nrow = 2,
                     byrow = TRUE,
                     dimnames = list(Gender = c("Male", "Female"),
                                     Product = c("A", "B")))
table_data

# Perform Chi-Square Test of Independence
chisq.test(table_data)
```


**Visualization Methods (Exploratory Data Analysis)**




 **--- Visualizing a Single Variable --- Histogram**

```r

hist(height,
     main = "Histogram of Heights",
     xlab = "Height (cm)",
     col = "lightblue",
     border = "black")

# Boxplot
boxplot(height,
        main = "Boxplot of Heights",
        ylab = "Height (cm)",
        col = "lightgreen")
```

**---Examining Multiple Variables ---Scatter Plot**

```r

plot(height, weight,
     main = "Scatter Plot of Height vs Weight",
     xlab = "Height (cm)",
     ylab = "Weight (kg)",
     col = "blue",
     pch = 19)

# Pairwise Scatter Plot (for more than 2 numeric variables)
data_matrix <- data.frame(height, weight)
pairs(data_matrix, main = "Pairwise Scatter Plot", pch = 19, col = "darkgreen")


```

**Iris Corpus**

```r
# Load CSV file (adjust the path if necessary)
iris <- read.csv("/content/Iris.csv")
# View the first few rows
head(iris)
```

```r
# Structure
str(iris)

```

```r
# summary
summary(iris)
```

```r
# Mean, Median, SD, Variance, Range
mean(iris$SepalLengthCm)
median(iris$SepalLengthCm)
sd(iris$SepalLengthCm)
var(iris$SepalLengthCm)
range(iris$SepalLengthCm)
```

```r
# Correlation between numeric features
cor(iris[,1:4])
```

```r
# Group-wise summary by Species
aggregate(. ~ Species, data = iris, FUN = mean)
```

**Histogram**

```r
hist(iris$SepalLengthCm,
     main = "Histogram of Sepal Length",
     xlab = "Sepal Length (cm)",
     col = "lightblue",
     border = "black")
```

**Boxplots**

```r
boxplot(SepalLengthCm ~ Species,
        data = iris,
        main = "Boxplot of Sepal Length by Species",
        xlab = "Species",
        ylab = "Sepal Length (cm)",
        col = c("lightblue", "lightgreen", "lightpink"))
```

**Scatter plots**

```r
plot(iris$SepalLengthCm, iris$PetalLengthCm,
     main = "Sepal Length vs Petal Length",
     xlab = "Sepal Length (cm)",
     ylab = "Petal Length (cm)",
     col = as.factor(iris$Species),
     pch = 19)
legend("topleft",
       legend = unique(iris$Species),
       col = 1:length(unique(iris$Species)),
       pch = 19)
```

**Correlation heatmap**

```r
cor_matrix <- cor(iris[, c("SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm")])
heatmap(cor_matrix,
        main = "Correlation Heatmap (Base R)",
        col = colorRampPalette(c("white", "blue"))(100),
        scale = "none")
```

**ANOVA (Testing Species Differences)**

```r
anova_model <- aov(SepalLengthCm ~ Species, data = iris)
summary(anova_model)
TukeyHSD(anova_model)
```
